# Concet

concet adalah menggabung 2 dataFrame by index/ menggabungkan secara index.

```python
data1 ={"country":["JPN", "CHN", "IND"],
        "year": [2019, 2019, 2020]}

data2 = {"country":["JPN", "CHN", "IND"],
         "city": ["Tokyo", "Beijing", "Seoul"]}
```

```python
data1 = pd.DataFrame(data1)
data2 = pd.DataFrame(data2)
```

```python
print(data1)
print(data2)
```

```python
# concet dengan menggabungkan data secara columns dengan axis = 1
df_kolom = pd.concat([data1, data2], axis = 1)
df_kolom
```

```python
# Concet dengan menggabungkan data dengan pada rows (baris) dengan axis = 0
df_baris = pd.concat([data1, data2], axis = 0)
df_baris
```

bisa di lihat dari tabel di atas bahwasannya terdapat nilai NaN sedangkan menggabungkan kolums secara tidak ada. itu alasannya sesui dengan isi data1 dan data2

# Table join

```python
population = pd.read_csv("/content/WorldBank_POP.csv")
```

```python
population.head()
```

```python
gpd = pd.read_csv("/content/WorldBank_GDP.csv")
```

```python
gpd.head()
```

```python
world_bank_outer = population.merge(gpd,
                                  on = ["Country Code","Country Name", "Year"],
                                  how = "outer")
world_bank_outer
```

```python
world_bank_inner = population.merge(gpd,
                                  on = ["Country Code","Country Name", "Year"],
                                  how = "inner")
world_bank_inner
```

```python
world_bank_left = population.merge(gpd,
                                  on = ["Country Code","Country Name", "Year"],
                                  how = "left")
world_bank_left
```

```python
world_bank_right = population.merge(gpd,
                                  on = ["Country Code","Country Name", "Year"],
                                  how = "right")
world_bank_right
```

```python
world_bank_multiple = population.merge(gpd,
                                  on = ["Country Code","Country Name", "Year"],
                                  )
world_bank_multiple
```
